```python
opstack = []

def opPop(): 
    return opstack.pop()

def opPush(value):
    opstack.append(value)
    return True
```

```python
dictstack = [{}]

def dictPop():
    return dictstack.pop()

# add new dictionary
def dictPush():
    dictstack.append({})
    return True

def define(name, value):
    # add definition on top dicitonary
    dictstack[-1][name] = value
    return True

def lookup(name):
    # reverse to look at top of stack when iterating
    dictstack.reverse()
    for d in dictstack:
        for k, v in d.items():
            # found
            if name == k:
                r = v
                dictstack.reverse()
                return r
    # revert back to it's normal self
    dictstack.reverse()
    return 
```

```python
def checkVar(value):
    value = str(value) 
    # find out if it's a variable or actual number
    if (value[0] == '-' and value[1:].isdigit()) or value.isdigit():
        value = int(value)
    elif value == 'false':
        return False
    elif value == 'true':
        return True
    else:
        value = lookup(value)
    return value

def add():    
    opPush(checkVar(opPop()) + checkVar(opPop()))
def sub():
    opPush(checkVar(opPop()) - checkVar(opPop()))
def mul():
    opPush(checkVar(opPop()) * checkVar(opPop()))
def div():
    opPush(checkVar(opPop()) / checkVar(opPop()))
def eq():
    opPush(checkVar(opPop()) == checkVar(opPop()))
def lt():
    opPush(checkVar(opPop()) > checkVar(opPop()))
def gt():
    opPush(checkVar(opPop()) < checkVar(opPop()))
```

```python
def psand():
    opPush(opPop() and opPop())
def psnot():
    opPush(not opPop())
def psor():
    opPush(opPop() or opPop())
```

```python
def dup():
    opPush(opstack[-1])
def exch():
    val1 = opPop()
    val2 = opPop()
    opPush(val1)
    opPush(val2)
def pop():
    opPop()
```

```python
def dictz():
    opPush({})
def begin():
    top = opPop()
    # only move dictionaries
    if type(top) is dict:
        dictPush()
        return True
    else:
        print("ERROR: Can't move non-dictionary type to dictionary stack.")
        return False
    
def end():
    # pop dictionary stack
    dictPop()
    
def psDef():
    var1 = opPop()
    var2 = opPop()
    # is var2 a variable name
    if type(var2) is not str:
        return False
    # add variable 
    define(var2, var1)
    return True
```

```python
def equals():
    print(opPop())
    
def stack():
    for item in opstack:
        print(item)
```

```python
import re

# checks for matching braces
def matchBraces(s):
    s= iter(s[0:])
    for c in s:
        if c == '{':
            if grouper(s):
                continue
            else:
                return False
        elif c == '}':
            print("Syntax error: missing a \'{0}\'".format('{'))
            return False
    return True
            
    
def grouper(it):
    for c in it:
        if c == '{':
            grouper(it)
        elif c == '}':
            return True
        prevWord = c
    print("Syntax error: missing a \'{0}\'".format('}'))
    return False

# tokenizes code and returns token array
def tokenize(s):
    l = re.findall("/?[a-zA-Z][a-zA-Z0-9_]*|[-]?[0-9]+|[}{]+|%.*|[^ \t\n]", s)
    # fix bug where a token ends up being '}{'
    for i in range(len(l)):
        if l[i] == '}{':
            l[i:i+1] = ['}', '{']
    
    if matchBraces(l):
        return l
    else:
        return False

# takes a list of tokens and returns a code array
def parse(tokens):
    return innerParse(iter(tokens[0:]))
            
def innerParse(it):
    l = []
    for c in it:
        if c == '}':
            return l
        elif c == '{':
            l.append(innerParse(it))
        else:
            l.append(c)
    return l

tokenize(
'''
/fact{ 
dictz begin
      /n exch def
         n 2 lt
         {1}
         {n -1 add fact n mul }
      ifelse
   end
}def { d}
5 fact =
'''
)
```

```python
def interpret(code): # code is a code array
    for c in code:
        if c == "def":
            psDef()
        elif c == "dictz":
            dictz()
        elif c == "begin":
            begin()
        elif c == "end":
            end()
        elif c == "stack":
            stack()
        elif c == "=":
            equals()
        elif c == "begin":
            begin()
        elif c == "pop":
            pop()
        elif c == "exch":
            exch()
        elif c == "dup":
            dup()
        elif c == "lt":
            lt()
        elif c == "gt":
            gt()
        elif c == "eq":
            eq()
        elif c == "add":
            add()
        elif c == "sub":
            sub()
        elif c == "mul":
            mul()
        elif c == "div":
            div()
        elif c == "not":
            psnot()
        elif c == "and":
            psand()
        elif c == "or":
            psor()
        elif c == "if":
            carr = opPop()
            if opPop():
                interpret(carr)
        elif c == "ifelse":
            second = opPop()
            first = opPop()
            if(opPop()):
                interpret(first)
            else:
                interpret(second)
        # name constants, numbers, booleans
        else:
            if type(c) is list:
                opPush(c)
            elif c[0] == '/':
                opPush(c[1:])
            else:
                c = checkVar(c)
                # code array
                if type(c) is list:
                    interpret(c)
                else:
                    opPush(c)
        

def interpreter(s): # s is a string
    interpret(parse(tokenize(s)))
```
